### Named Entity Recognition


```python
sentence="The Eiffel Tower stands on four lattice-girder piers that taper inward and join to form a single large vertical tower. As they curve inward, the piers are connected to each other by networks of girders at two levels that afford viewing platforms for tourists. By contrast, the four semicircular arches at the tower’s base are purely aesthetic elements that serve no structural function. "
```

```python
import nltk
words=nltk.word_tokenize(sentence)
```

```python
tag_elements=nltk.pos_tag(words)
```

```python
import nltk
nltk.download('maxent_ne_chunker_tab')
nltk.download('words')
```

```python
nltk.ne_chunk(tag_elements)
```
